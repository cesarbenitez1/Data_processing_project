```python
import pandas as pd
import matplotlib.pyplot as plt

# Asegrate de cargar tus datos o ajustar segn tu estructura de cdigo
datos = pd.read_csv("datos_procesados.csv")

# Crear subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Grfica de torta para la cantidad de anmicos
anemia_counts = datos['anaemia'].value_counts()
axs[0, 0].pie(anemia_counts, labels=anemia_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
axs[0, 0].set_title('Cantidad de Anmicos')

# Grfica de torta para la cantidad de diabticos
diabetes_counts = datos['diabetes'].value_counts()
axs[0, 1].pie(diabetes_counts, labels=diabetes_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightyellow'])
axs[0, 1].set_title('Cantidad de Diabticos')

# Grfica de torta para la cantidad de fumadores
smoking_counts = datos['smoking'].value_counts()
axs[1, 0].pie(smoking_counts, labels=smoking_counts.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue'])
axs[1, 0].set_title('Cantidad de Fumadores')

# Grfica de torta para la cantidad de muertos
death_counts = datos['DEATH_EVENT'].value_counts()
axs[1, 1].pie(death_counts, labels=death_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
axs[1, 1].set_title('Cantidad de Muertos')

# Ajustar el diseo y mostrar la grfica
plt.tight_layout()
plt.show()

```
